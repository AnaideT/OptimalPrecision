```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
m = 4
beta = 3.9
```

```python
timestep = 1
```

```python
x = timestep*np.arange(120//timestep)
vitesse = 1/beta*(1-np.exp(-x*beta/m))
position = x/beta-m/(beta**2)*(1-np.exp(-x*beta/m))
```

```python
y = timestep*np.zeros(120//timestep)
z = timestep*np.zeros(120//timestep)
```

```python
for j in np.arange(50//timestep+1):
    y[j+20//timestep]=vitesse[j]
```

```python
for j in np.arange(50//timestep+1):
    z[j+20//timestep]=position[j]
```

```python
#plt.plot(x, y)

z[(70//timestep+1):(120//timestep)]=z[70//timestep]
  
xx = [50,50]
yx = [15,-2]

fig_width = 15
fig, ax = plt.subplots(1, 1, figsize=(fig_width, fig_width/1.618))
#N_show = 10
#from cycler import cycler
#ax.set_prop_cycle(cycler('color', [plt.cm.magma(h) for h in np.linspace(0, 1, N_show)]))
#ax.plot(time, a[::(niter//N_show), :].T, '-', alpha=.5, lw=1)
ax.set_title(r'Position $x$')
ax.set_xlabel('Time (ms)', fontsize=14)
ax.set_ylabel('x ', fontsize=14)
#ax.set_xlim(0, 150)

      
ax.plot(x-20, z)

ax.plot(xx, yx, "--", label="ligne --")


plt.show()
```

Orange: position evolution.  
Blue: speed evolution.  
Green (dashed line): time when the target is pointed.
