### Other notebooks

Les autres notebooks sont soit des projets n'ayant pas entièrement abouti, soit, plus généralement, la version définitive des notebooks (notamment car le code a été grandement accéléré par la façon de l'écrire).  
  
2017-05-19 Tables of parameters.ipynb  >>>  ce notebook avait pour objectif de donner les valeurs optimales des paramètres de convergence suivant la valeur du pas de temps $dt$ utilisé dans le modèle de H&W, ainsi que des indicateurs sur la vitesse de convergence.  
  
2017-05-31 Automation of the learning process.ipynb  >>>  ce notebook est le notebook d'appel du script implémenté par le fichier HarrisWolpertandPursuit.py (ainsi que des fonctions pour les plots) ; il faut importer dans le notebook les fonctions présentes et utilisées dans le script, puis lancer un appel au notebook avec les paramètres que l'on veut (si le modèle a déjà été testé avec ces paramètres, la fonction va chercher les informations dans les pickle), ensuite on peut visionner les résultats. Le notebook fournit également une comparaison visuelle avec les modèles de bangbang  
  
2017-06-01 Bang-bang solution.ipynb  >>>  notebook définitif sur la solution bangbang symétrique, étendu à la poursuite  
  
2017-07-10 Asymmetrical bang-bang.ipynb  >>>  notebook non définitif car étendu à la poursuite via le notebook suivant  
  
2017-07-11 Asymmetrical bang-bang and pursuit.ipynb  >>>  notebook définitif sur la solution bangbang asymétrique, étendu à la poursuite  
  
2017-07-10 Saccadic correction.ipynb  >>>  modèle de correction saccadique fondé sur le modèle de H&W (à voir si utilisable pour l'adaptation saccadique)  
  
2017-06-13 Generalized coordinates.ipynb  >>>  modèle définitif de H&W étendu à la poursuite. Il permet le recours aux coordonnées généralisées (à voir quels sont les paramètres à modifier pour le cas strictement supérieur à deux coordonnées) ; il permet la poursuite de trajectoires non rectilignes (sinusoïdales par exemple) à condition que celles-ci soient des connaissances a priori (possible de rendre cela légitime par le biais d'un délai et d'une trajectoire cyclique : voir _Transitioning from Saccade to Smooth Pursuit Eye Movements using Linear Quadratic Tracking Control_ Korentis and Enderle, J Bioengineer & Biomedical Sci 2016, 6:3)  
  
2017-07-13 Noised trajectories.ipynb  >>>  analyse graphique de "vraies" trajectoires (bruitées) (moyenne théorique + moyenne empirique + intervalle de confiance) + possibilité de récupérer le Control Signal à partir de la position ou la vitesse dans le cadre du modèle de H&W
  
2017-07-26 Anticipatory SPEM .ipynb  >>>  étude de l'anticipation avant la poursuite d'un mobile. Le cadre de l'expérience : on fixe un point au centre de l'écran, après un délai de fixation, le point va à droite avec une probabilité $p$ et à gauche avec une probabilité $1-p$. Le sujet doit suivre le point et cela recommence. Plus $p$ est proche de 1, plus le sujet aura tendance à anticiper la trajectoire.
