# Condition  2 : independance

```python
import numpy as np

# creating the condition 2 : independance between the probabilities of apparition of Hand and Eye for left and right
'''
Probabilities of apparition for both the hand and eye targets:
1 :
    LL = .04
    Hand left, eye left
2 :
    LR = .16
    Hand left, eye right
3 :
    RL = .16
    Hand right, eye left
4 :
    RR = .64
    Hand right, eye right
    
(with R = 80 and L = 20)
'''

# p = np.zeros((2, 2))
# p = [[LL, LR], [RL, RR]]

# number of trials per session
n = 40 

p_array = []
    

for i in range(1, n+1):
    n = np.random.rand()
    if  n <= 0.04: # knowing LL = .04
        p = 'sequence1'
    elif n > 0.04 and n <= 0.20: # knowing LR = .16
        p = 'sequence2'
    elif n > 0.20 and n <= 0.36: # knowing RL = .16
        p = 'sequence3'
    else: # knowing RR = .64
        p = 'sequence4'
    p_array.append(p)
    print(p)

print(p_array)

a = p_array.count('sequence1')
b = p_array.count('sequence2')
c = p_array.count('sequence3')
d = p_array.count('sequence4')

print(a/n, b/n, c/n, d/n)
    
```
