# Condition  2 : independance

```python
import numpy as np

# creating the condition 2 : independance between the probabilities of apparition of Hand and Eye for left and right
'''
Probabilities of apparition for both the hand and eye targets:
1 :
    LL = .04
    Hand left, eye left
2 :
    LR = .16
    Hand left, eye right
3 :
    RL = .16
    Hand right, eye left
4 :
    RR = .64
    Hand right, eye right
    
(with R = 80 and L = 20)
'''

# p = np.zeros((2, 2))
# p = [[LL, LR], [RL, RR]]

def get_number():
    n = np.random.rand()
    if  n <= 0.04: # knowing LL = .04
        return 'sequence1'
    elif n > 0.04 and n <= 0.20: # knowing LR = .16
        return 'sequence2'
    elif n > 0.20 and n <= 0.36: # knowing RL = .16
        return 'sequence3'
    else: # knowing RR = .64
        return 'sequence4'

for i in range(1, 41):
    print(get_number())
```
