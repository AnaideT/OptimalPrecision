# Condition 1 : baseline

```python
import numpy as np

# Condition 1 : baseline (40 trials)
'''
Probabilities of apparition for both the hand and eye targets:
1 :
    LL = .25
    Hand left, eye left
2 :
    LR = .25
    Hand left, eye right
3 :
    RL = .25
    Hand right, eye left
4 :
    RR = .25
    Hand right, eye right
'''


# p = np.zeros((2, 2))
# p = [[LL, LR], [RL, RR]]

# number of trials per session
n = 40

p_array =[]

for i in range(1, n+1):
    n = np.random.rand()
    if  n <= 0.25: # knowing LL = .25
        p = 'sequence1'
    elif n > 0.25 and n <= 0.50: # knowing LR = .25
        p = 'sequence2'
    elif n > 0.50 and n <= 0.75: # knowing RL = .25
        p = 'sequence3'
    else: # knowing RR = .25
        p = 'sequence4'
    p_array.append(p)
    print(p)

print(p_array)
a = p_array.count('sequence1')
b = p_array.count('sequence2')
c = p_array.count('sequence3')
d = p_array.count('sequence4')

print(a, b, c, d)

print(a/n, b/n, c/n, d/n)



```
